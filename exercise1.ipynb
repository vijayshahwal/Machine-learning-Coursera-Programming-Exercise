```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
data=pd.read_csv("ex1\ex1data1.txt",names=["X","y"])
m=len(data.y)
x=np.array(data.X)
y=np.array(data.y)
```

```python
plt.style.use("seaborn")
plt.scatter(x,y)
plt.show()
```

```python
ones=np.ones_like(x)
X=np.column_stack((ones,x))
theta=np.zeros(2)
```

```python
iteration=1500
alpha=0.01
```

```python
def warmUpExercise():
    return np.eye(5)
```

```python
def computeCost(X,y,theta):
    m=len(y)
    J = (np.sum((np.dot(X,theta) - y)**2))/(2*m)
    return J
```

```python
computeCost(X,y,theta)
```

```python
def gradientDescent(X,y,theta,alpha,num_iter):
    m=len(y)
    J_history=np.zeros(num_iter)
    for i in range(num_iter):
        theta=theta-(alpha/m)*np.sum((np.dot(X,theta)-y)[:,None]*X,axis=0)
        J_history[i]=computeCost(X,y,theta)
    return theta,J_history
```

```python
theta,hist=gradientDescent(X,y,theta,alpha,iteration)
```

```python
plt.style.use("fivethirtyeight")
plt.scatter(x,y)
plt.plot(X,y,"rx",x,np.dot(X,theta),color="red")
plt.show()
```

```python
predict1=np.dot([1,3.5],theta)
print("population profit of 35,000",predict1*10000)
predict2=np.dot([1,7],theta)
print("population profit of 70,000",predict2*10000)
```

```python
theta0_vals=np.linspace(-10,10,100)
theta1_vals=np.linspace(-1,4,100)
```

```python
J_vals=np.zeros((len(theta0_vals),len(theta0_vals)))
```

```python
for i in range(len(theta0_vals)):
    for j in range(len(theta1_vals)):
        t=np.array([theta0_vals[i],theta1_vals[j]])
        J_vals[i][j]=computeCost(X,y,t)
```

```python
fig = plt.figure()
ax = plt.subplot(111,projection='3d')
Axes3D.plot_surface(ax,theta0_vals,theta1_vals,J_vals,cmap=cm.coolwarm)
plt.show()
```
